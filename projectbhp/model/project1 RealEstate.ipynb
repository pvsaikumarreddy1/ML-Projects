```python
df2=df1.drop(['availability','society','area_type','balcony'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2=df2.dropna()
```

```python
df2
```

```python
df2['BHK']=df2['size'].apply(lambda x: int(x.split(' ')[0]))
df2.head()
```

```python
df2=df2.drop('size',axis=1)
```

```python
df2.head()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return True
    return False
```

```python
df2[df2['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df3=df2.copy()
df3

```

```python
df3.total_sqft = df2.total_sqft.apply(convert_sqft_to_num)
df3.head(30)

```

```python
df3.isnull().sum()
```

```python
df4 = df3[df3.total_sqft.notnull()]
df4.head()
```

```python
len(df4.location.unique())
```

```python
df4['price_per_squareft']=df4['price']*100000/df4['total_sqft']
df4
```

```python
location_stats=df4.location.value_counts()
location_stats
```

```python
location_stats[location_stats<=10]
```

```python
location_stats_less_than_10=location_stats[location_stats<=10]
```

```python
df4.location=df4.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
```

```python
df4.head(30)
```

```python
df4[df4.total_sqft/df4.BHK<300].head()
```

```python
df4.shape
```

```python
df5=df4[~(df4.total_sqft/df4.BHK<300)]
df5.shape

```

```python
df5.head()
```

```python
def remove_outliers(df):
    df_out=pd.DataFrame()
    for key,i in df.groupby('location'):
        mean=np.mean(i.price_per_squareft)
        std=np.std(i.price_per_squareft)
        reduced_df=i[(i.price_per_squareft>(mean-std)) & (i.price_per_squareft<=(mean+std))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

```

```python
df6=remove_outliers(df5)
df6.shape
```

```python
df7=df6[df6.bath<df6.BHK+2]
df7.shape
```

```python
df7.head(30)
```

```python
df7.location.value_counts()
```

```python
dummies=pd.get_dummies(df7.location)
dummies.head(3)
```

```python
df8=pd.concat([df7,dummies.drop('other',axis=1)],axis=1)
df8.head(3)
```

```python
df8.shape
```

```python
df9=df8.drop(['location','price_per_squareft'],axis=1)
df9.shape
```

```python
X=df9.drop('price',axis=1)
y=df9.price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
lin_reg.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=10, test_size=0.3, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python

from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lin_reg.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1250,2,2)
```
